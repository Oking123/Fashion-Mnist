```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plot_model(file, name):
    loss_data = pd.read_csv(file)
    plt.title("{} loss".format(name))
    plt.plot(loss_data['epochs'],loss_data['loss'],label='train',color='black',linestyle='--',linewidth=1)
    plt.plot(loss_data['epochs'],loss_data['val_loss'],label='test',color='black',linestyle='-',linewidth=1)
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel('loss')
    plt.show()
    plt.title("{} Accuracy".format(name))
    plt.plot(loss_data['epochs'],loss_data['acc'],label='train',color='black',linestyle='--',linewidth=1)
    plt.plot(loss_data['epochs'],loss_data['val_acc'],label='test',color='black',linestyle='-',linewidth=1)
    plt.legend()
    plt.xlabel('epochs')
    plt.ylabel('accuracy')
    plt.show()
    plt.plot(loss_data['loss'], loss_data['acc'],label='train',color='black',linestyle='--',linewidth=1)
    plt.scatter(loss_data['val_loss'], loss_data['val_acc'],label='test', marker = 'x',linewidth=1,color='black')
    plt.title("{} Acc vs Loss".format(name))
    plt.xlabel('loss')
    plt.ylabel('acc')
    plt.legend()
    plt.xlim((0.2, 0.5))
    plt.ylim((0.8, 0.95))
    plt.show()
plot_model('loss_graph_model3.csv', 'VGG like model')
```

```python
plot_model("loss_graph_model2.csv","Two layer")
```

```python
plot_model("loss_graph_dnn.csv","DNN")
```

```python
plot_model("loss_graph_1_layer.csv","One-layer")
```
